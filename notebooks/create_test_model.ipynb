```python
# Import dependencies
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score
from sklearn.discriminant_analysis import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
# #load and prep the data
data_file_path = 'pokemon.csv'
data_df = pd.read_csv(data_file_path)
```

```python
#filter to just the numeric data and get rid of what's not needed
data_df = data_df.loc[data_df['base_experience'] >0 ]
filtered_df = data_df[data_df.describe().columns.to_numpy()]
del filtered_df['ability_2_id']
del filtered_df['ability_1_id']
del filtered_df['type_1_id']
del filtered_df['id']

# filtered_df = filtered_df.fillna(0)
filtered_df = filtered_df.dropna()
filtered_df
```

```python
#see how many types exist
num_types = len(filtered_df['type_0_id'].value_counts())
```

```python
y = filtered_df['type_0_id']
X = filtered_df
del X['type_0_id']
X
```

```python
model= KMeans(n_clusters=int(num_types))
model.fit(filtered_df)
```

```python
# Make predictions about the data clusters using the trained model
predictions = model.predict(filtered_df)

filtered_df['Kmeans'] = predictions
filtered_df['OG_type'] = y
```

```python
accuracy = accuracy_score(y, predictions)
print("Accuracy:", accuracy)
```
